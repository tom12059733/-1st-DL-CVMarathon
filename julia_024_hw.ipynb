# Julia 基礎圖表繪製：圖表結合敘述統計

## 作業 024：繪製帶統計資訊的圖表

繪製 iris 資料的散佈圖，x 軸為 PetalLength，y 軸為 PetalWidth，並依不同花的種類分別算出 PetalLength 及 PetalWidth 的平均值，將平均值以鉛直線與水平線標在圖上。（請完成以下程式碼）

```julia
using StatsBase, RDatasets
using Gadfly
```

```julia
iris = dataset("datasets", "iris")
```

```julia
p = plot(iris, x="PetalLength", y="PetalWidth", Geom.point)
```

```julia
typeof(iris)
```

```julia
by(iris, :Species, :PetalLength => mean, :PetalWidth => mean)
```

```julia
p = plot(iris, x="PetalLength", y="PetalWidth",
    xintercept=[mean(iris[!, :PetalLength])],  
    yintercept=[mean(iris[!, :PetalWidth])], 
    Geom.vline(color=["black"]),  
    Geom.hline(color=["red"]),
    color="Species",
    Geom.point)
```
